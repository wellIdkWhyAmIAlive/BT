```python
import hashlib
from datetime import datetime
```

```python
class SimpleBlock:
    def __init__(self, block_no, payload, prev_hash):
        self.block_no = block_no
        self.payload = payload
        self.prev_hash = prev_hash
        self.created_at = datetime.utcnow()
        self.block_hash = self.generate_hash()

    def generate_hash(self):
        content = f"{self.block_no}{self.payload}{self.prev_hash}{self.created_at}"
        return hashlib.sha256(content.encode()).hexdigest()
```

```python
class SimpleBlockchain:
    def __init__(self):
        self.ledger = []
        self.initialize_chain()

    def initialize_chain(self):
        genesis = SimpleBlock(0, "Initial Block", "None")
        self.ledger.append(genesis)

    def insert_block(self, payload):
        last_block = self.ledger[-1]
        new_block = SimpleBlock(
            block_no=len(self.ledger),
            payload=payload,
            prev_hash=last_block.block_hash
        )
        self.ledger.append(new_block)

    def verify_chain(self):
        for i in range(1, len(self.ledger)):
            current = self.ledger[i]
            previous = self.ledger[i - 1]

            if current.block_hash != current.generate_hash():
                return False

            if current.prev_hash != previous.block_hash:
                return False

        return True
```

```python
# Create blockchain object
chain = SimpleBlockchain()

# Add blocks
chain.insert_block("User X sent 5 tokens")
chain.insert_block({"sender": "A", "receiver": "B", "value": 25})
chain.insert_block(["TX100", "TX101", "TX102"])

# Display blocks
for blk in chain.ledger:
    print("\n-------------------------")
    print("Block Number:", blk.block_no)
    print("Timestamp:", blk.created_at)
    print("Data:", blk.payload)
    print("Previous Hash:", blk.prev_hash)
    print("Current Hash:", blk.block_hash)

# Validate blockchain
print("\nBlockchain Validity:", chain.verify_chain())
```
